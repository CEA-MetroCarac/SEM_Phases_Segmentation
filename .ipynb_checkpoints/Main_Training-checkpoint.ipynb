# Training and testing U-Net Model

## Load Training/Validation dataset

Images are loaded for training purposes.
This cell is specifically used to load the images and split them into training and validation datasets. The `val_perc` parameter defines the proportion of the data used for validation, while the remaining `1-val_perc` is used for training. 

- `Patch_size`  
  This defines the size (in pixels) of square patches that the input images are divided into.  
  For example, if `Patch_size` is $256$, each image will be processed in blocks of $256 \times 256$ pixels.  
  This is commonly used in patch-based training for deep learning models, particularly in image segmentation tasks.

- `n_classes`  
  This indicates the total number of classes to be predicted in the image.  

<div class="alert alert-block alert-info">
<b>Training_param is a training flag.</b> When set to True, the model is trained. If set to False, the function attempts to load a pre-trained model with the specified name (`Save_name`). If the models' name does not exist, an error is raised and a valid model name is required.
</div>


```python
Patch_size = 256
Data_folder = 'Data'
n_classes = 4
Training_param = False
# Load training/Validation data set

if Training_param == True:
    Training_im_path = Path(Data_folder) / f'Augmented_Images_{Patch_size}'
    Training_lb_path = Path(Data_folder) / f'Augmented_Labels_{Patch_size}'
    
    X_train, y_train, y_train_cat, X_val, y_val, y_val_cat, _, _, _ = Load_and_prepare_data(
        for_training=True,
        data_dir=(Training_im_path,Training_lb_path), patch_size=Patch_size, 
        val_perc = 0.2, n_classes=n_classes, 
        normalize_images=True, 
        shuffle=True
    )

else:
    X_train = None 
    y_train_cat = None 
    X_val = None 
    y_val_cat = None
    
```

## Train new model or Load already trained model
Two segmentation techniques are available:

1. Direct U-Net segmentation: A single U-Net model is either trained or loaded and used to perform direct segmentation. The final output is a single prediction map.

2. Ensemble method: This technique involves training nb_models separate U-Net models, each on a different split of the dataset (randomly shuffled between training and validation for each model). During prediction, all models produce segmentation maps for the same image. These maps are then averaged to generate a final probability map. This method typically yields more robust and generalized results, but is more computationally intensive and time-consuming.

```python
blur_accuracy_percent = np.zeros(X_test.shape[0])

for i in range(y_test.shape[0]):
    target = y_test[i, :, :].reshape((256,256))
    accuracy = calculate_accuracy(target, y_pred_argmax_test[i, :, :])
    blur_accuracy_percent[i] = accuracy*100  

# Element-wise difference
target = y_test[:,:,:,0]
prediction = y_pred_argmax_test
diff = target - prediction

# Mean Absolute Error (MAE) over all 28 elements
mae = np.mean(np.abs(diff))

# Root Mean Squared Error (RMSE) over all 28 elements
rmse = np.sqrt(np.mean(diff**2))

# Std

ecart = np.std(np.abs(diff))

# print("Difference matrix (target - prediction):\n", diff)
print("Mean Absolute Error (MAE):", mae)
print("Standard deviation (std):", ecart)
print("Root Mean Squared Error (RMSE):", rmse)
print(f'Accuracy percentage for U-Net = {np.mean(blur_accuracy_percent):.2f}%')
```

#### Confusion matrix

```python
class_names = ['Porosity', 'NiO', 'Ni', 'YSZ']

# Simulated ground truth and predictions (replace these with your actual data)
# Assuming y_test and y_pred_argmax_test are 2D arrays of shape (256, 256)
size = (256, 256)
Y_test = y_test[:,:,:,0]

# Compute Mean IoU
IOU_keras = tf.keras.metrics.MeanIoU(num_classes=n_classes)
IOU_keras.update_state(Y_test.flatten(), y_pred_argmax_test.flatten())
print("Mean IoU =", IOU_keras.result().numpy())

# Extract confusion matrix
confusion_matrix = IOU_keras.total_cm.numpy()

# Calculate IoU for each class
class_IoUs = []
for i, class_name in enumerate(class_names):
    intersection = confusion_matrix[i, i]
    union = (np.sum(confusion_matrix[i, :]) + np.sum(confusion_matrix[:, i]) - intersection)
    class_IoU = intersection / union if union != 0 else 0
    class_IoUs.append(class_IoU)
    print(f"IoU for {class_name}: {class_IoU}")

# Function to plot the confusion matrix
def plot_confusion_matrix(cm, classes, normalize=True, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], '.2f'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True Classes')
    plt.xlabel('Predicted label')
    plt.savefig('Figure/Confusion_matrix_Direct.tiff', format='tiff', dpi=300, bbox_inches='tight')
    
# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(confusion_matrix, classes=class_names, normalize=True,
                      title='Normalized Confusion Matrix')
plt.show()
```

```python
Testing_im_path = Path(Data_folder) / f'Testing_Images_{Patch_size}'
Testing_lb_path = Path(Data_folder) / f'Testing_Labels_{Patch_size}'

_, _, _, _, _, _, X_test, y_test, y_test_cat = Load_and_prepare_data(
    for_training=False,
    data_dir=(Testing_im_path,Testing_lb_path), patch_size=Patch_size, 
    val_perc=None, n_classes=n_classes, 
    normalize_images=True, 
    shuffle=False
)

def Ensemble_prediction(X_test, X_train, y_train_cat, X_val, y_val_cat, Num_of_models = 10):
    # Predict for each model
    stacked = []
    
    for i in range(Num_of_models):
        print('Loading model ', i)

        Save_name      = 'UNet_Norm_256px_1.5kV_'+str(i)
        model_folder   = 'Ensemble_models/'
        history_folder = 'Ensemble_history/'
        
        model,history = Main_model(X_train, y_train_cat, X_val, y_val_cat,
                                   Save_name, model_folder, history_folder,
                                   Training_param = False, Summary_param = False,
                                   batch_size = 32, epochs = 100, learning_rate = 1e-3)
    
    
        y_pred_test = model.predict(X_test, verbose = 0)
        stacked.append(y_pred_test)
    
    stacked_array  = np.stack(stacked, axis=0) 
    avg_prediction = np.mean(stacked_array, axis=0)
    avg_y_pred_argmax_test = np.argmax(avg_prediction, axis=-1)
    avg_proba_map_pred = np.max(avg_prediction, axis=-1)

    return avg_prediction, avg_y_pred_argmax_test, avg_proba_map_pred

avg_prediction, avg_y_pred_argmax_test, avg_proba_map_pred = Ensemble_prediction(
    X_test, X_train, y_train_cat, X_val, y_val_cat, 
    Num_of_models = Num_of_models
)

# Load images and predict
range_max = X_test.shape[0]
num_values = 10
random_vector = np.random.choice(range_max, num_values, replace=False)
rows = len(random_vector)

fig, axs = plt.subplots(rows, 5, figsize=(15, 5 * rows))  # Adjust figsize dynamically

def calculate_accuracy(gt, pred):
    """Calculate the pixel accuracy between ground truth and prediction."""
    correct = np.sum(gt == pred)
    total = gt.size
    return correct / total

c = 0
for i in random_vector:  # Iterate only over the limited number of rows
    target = y_test[i, :, :].reshape((Patch_size,Patch_size))
    accuracy = calculate_accuracy(target, avg_y_pred_argmax_test[i, :, :])
    accuracy_percent = accuracy*100  # Convert to percentage
    
    # First subplot for the SEM image
    axs[c, 0].imshow(X_test[i, :, :], cmap='gray')
    axs[c, 0].set_title('SEM image')
    axs[c, 0].axis('off')  # Hide the axis
    
    # Second subplot for the ground truth
    y_T = y_test[i, :, :]
    axs[c, 1].imshow(y_T[:,:,0], cmap='gray')
    axs[c, 1].set_title('Target')
    axs[c, 1].axis('off')  # Hide the axis
    ground_truth_image = (85 * y_T[:,:,0]).astype(np.uint8)
    tiff.imwrite('Figure/Prediction/' +f'Target_ensemble_{i}.tif', ground_truth_image)
    
    # Third subplot for the prediction
    axs[c, 2].imshow(avg_y_pred_argmax_test[i, :, :], cmap='gray')
    title = f'Prediction - Accuracy: {accuracy_percent:.2f}%'
    axs[c, 2].set_title(title)
    axs[c, 2].axis('off')  # Hide the axis
    prediction_image = np.clip(85 *  avg_y_pred_argmax_test[i, :, :], 0, 255).astype(np.uint8)
    tiff.imwrite('Figure/Prediction/' +f'Prediction_ensemble_{i}.tif', prediction_image)

    # Fourth subplot for the prediction
    Diff = np.abs(avg_y_pred_argmax_test[i, :, :]-y_T[:,:,0])
    axs[c, 3].imshow(Diff, cmap='gray')
    axs[c, 3].set_title('Difference')
    axs[c, 3].axis('off')  # Hide the axis
    Diff = np.clip(85 *  Diff, 0, 255).astype(np.uint8)
    tiff.imwrite('Figure/Prediction/' +f'Diff_ensemble_{i}.tif', Diff)

    # Fith subplot for the prediction
    # Create a custom colormap
    cmap = plt.cm.RdYlGn
    red_yellow_green_cmap = LinearSegmentedColormap.from_list("RedYellowGreen", ["red", "yellow", "green"])

    axs[c, 4].imshow(avg_proba_map_pred[i, :, :], cmap=red_yellow_green_cmap, interpolation='nearest')
    im = axs[c, 4].imshow(avg_proba_map_pred[i, :, :], cmap=red_yellow_green_cmap, interpolation='nearest')
    cbar = fig.colorbar(im, ax=axs[c, 4], fraction=0.046, pad=0.04)
    cbar.ax.tick_params(labelsize=10)
    axs[c, 4].axis('off')  # Hide the axis
    extent = axs[c, 4].get_window_extent().transformed(fig.dpi_scale_trans.inverted())  # Get the bounds of the subplot
    fig.savefig('Figure/Prediction/' +f'Proba_map_ensemble_{i}.tif', format='tiff', dpi=600, bbox_inches=extent, pad_inches=0)
    c = c+1
    
# Display the plot
plt.tight_layout()
plt.show()
```

```python
class_names = ['Porosity', 'NiO', 'Ni', 'YSZ']

# Simulated ground truth and predictions (replace these with your actual data)
# Assuming y_test and y_pred_argmax_test are 2D arrays of shape (256, 256)
size = (256, 256)
Y_test = y_test[:,:,:,0]

# Compute Mean IoU
IOU_keras = tf.keras.metrics.MeanIoU(num_classes=n_classes)
IOU_keras.update_state(Y_test.flatten(), avg_y_pred_argmax_test.flatten())
print("Mean IoU =", IOU_keras.result().numpy())

# Extract confusion matrix
confusion_matrix = IOU_keras.total_cm.numpy()
print("Confusion Matrix:\n", confusion_matrix)

# Calculate IoU for each class
class_IoUs = []
for i, class_name in enumerate(class_names):
    intersection = confusion_matrix[i, i]
    union = (np.sum(confusion_matrix[i, :]) + np.sum(confusion_matrix[:, i]) - intersection)
    class_IoU = intersection / union if union != 0 else 0
    class_IoUs.append(class_IoU)
    print(f"IoU for {class_name}: {class_IoU}")

# Function to plot the confusion matrix
def plot_confusion_matrix(cm, classes, normalize=True, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], '.2f'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True Classes')
    plt.xlabel('Predicted label')
    plt.savefig('Figure/Confusion_matrix_ensemble.tiff', format='tiff', dpi=300, bbox_inches='tight')
    
# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(confusion_matrix, classes=class_names, normalize=True,
                      title='Normalized Confusion Matrix')
plt.show()
```

# Prediction Variance using Ensemble method

```python
Testing_im_path = Path(Data_folder) / f'Testing_Images_{Patch_size}'
Testing_lb_path = Path(Data_folder) / f'Testing_Labels_{Patch_size}'

_, _, _, _, _, _, X_test, y_test, y_test_cat = Load_and_prepare_data(
    for_training=False,
    data_dir=(Testing_im_path,Testing_lb_path), patch_size=Patch_size, 
    val_perc=None, n_classes=n_classes, 
    normalize_images=True, 
    shuffle=False
)

def Ensemble_prediction(X_test, X_train, y_train_cat, X_val, y_val_cat, Num_of_models = 10):
    # Predict for each model
    stacked = []
    
    for i in range(Num_of_models):
        print('Loading model ', i)

        Save_name      = 'UNet_Norm_256px_1.5kV_'+str(i)
        model_folder   = 'Ensemble_models/'
        history_folder = 'Ensemble_history/'
        
        model,history = Main_model(X_train, y_train_cat, X_val, y_val_cat,
                                   Save_name, model_folder, history_folder,
                                   Training_param = False, Summary_param = False,
                                   batch_size = 32, epochs = 100, learning_rate = 1e-3)
    
    
        y_pred_test = model.predict(X_test, verbose = 0)
        stacked.append(y_pred_test)
    
    stacked_array  = np.stack(stacked, axis=0) 
    
    avg_prediction = np.mean(stacked_array, axis=0)
    avg_y_pred_argmax_test = np.argmax(avg_prediction, axis=-1)
    avg_proba_map_pred = np.max(avg_prediction, axis=-1)

    return avg_prediction, avg_y_pred_argmax_test, avg_proba_map_pred


avg_prediction, avg_y_pred_argmax_test, avg_proba_map_pred = Ensemble_prediction(
    X_test, X_train, y_train_cat, X_val, y_val_cat, 
    Num_of_models = Num_of_models
)
```

```python
# Assume X_test is defined elsewhere
num_models = 10
num_test_images = X_test.shape[0]
height, width = X_test.shape[1], X_test.shape[2]

predictions = np.zeros((num_models, num_test_images, height, width))
probability_maps = np.zeros((num_models, num_test_images, height, width))

# Load models and make predictions
for i in range(num_models):
    print(f"Loading model {i+1}/{num_models}")

    Save_name      = f"UNet_Norm_{Patch_size}px_1.5kV_{i}"
    model_folder   = 'Ensemble_models/'
    history_folder = 'Ensemble_history/'
    
    model, _ = Main_model(
        X_train, y_train_cat, X_val, y_val_cat,
        Save_name, model_folder, history_folder,
        Training_param=False, Summary_param=False,
        batch_size=32, epochs=100, learning_rate=1e-3
    )
    
    y_pred_test = model.predict(X_test, verbose=1)
    predictions[i] = np.argmax(y_pred_test, axis=-1)
    probability_maps[i] = np.max(y_pred_test, axis=-1)

# Compute uncertainty maps
uncertainty_maps = np.std(probability_maps, axis=0)  # (num_test_images, height, width)

# Save uncertainty maps
output_dir = "Figure/Prediction/"
os.makedirs(output_dir, exist_ok=True)

for idx, uncertainty_map in enumerate(uncertainty_maps):
    output_path = f"{output_dir}/uncertainty_map_{idx}.tif"
    
    fig, ax = plt.subplots()
    im = ax.imshow(uncertainty_map, cmap='jet', interpolation='nearest')
    ax.axis('off')
    extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    plt.savefig(output_path, format='tiff', dpi=600, bbox_inches=extent, pad_inches=0)
    plt.close(fig)
    
    print(f"Saved uncertainty map for test image {idx+1} to {output_path}")
```
